```python
from firecrawl.firecrawl import FirecrawlApp
from dotenv import load_dotenv
import os
import pickle

load_dotenv()

api_key = os.getenv('FIRECRAWL_API')

app = FirecrawlApp(api_key=api_key)
```

```python
def scrape(url):
    scrape_status = app.scrape_url(
    url, 
    params={'formats': ['markdown']}
    )
    return scrape_status
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://daihoc.fpt.edu.vn/chuong-trinh-dao-tao/"

headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, "html.parser")

links = []
for a_tag in soup.find_all("a", href=True):  
    if "nganh-hoc" in a_tag["href"]:
        links.append(a_tag["href"])

for link in links:
    print(link)
```

```python
urls = ["https://daihoc.fpt.edu.vn/gioi-thieu-dai-hoc-fpt/",
        "https://daihoc.fpt.edu.vn/chuong-trinh-dao-tao/",
        ]
```

```python
for i in range(len(links)):
    links[i] = 'https://daihoc.fpt.edu.vn' + links[i]
```

```python
urls += links
```

```python
urls
```

```python
import time
data = []
for link in urls:
    print(link)
    scraped = scrape(link)
    time.sleep(10)
    data.append(scraped)
```

```python
data
```

```python
with open("crawl2.pkl", "wb") as f:
    pickle.dump(data, f)
```

```python
from IPython.display import display, Markdown
for i in data:
    display(Markdown(i['markdown']))
```

```python
with open("crawl2.pkl", "rb") as f:
    data = pickle.load(f)
```

```python
for url in data:
    for k,v in url['metadata'].items():
        if k == "url":
            url['url'] = v 
    del url['metadata']
```

```python
for dictionary in data:
    name = dictionary['url'].rstrip("/").split("/")[-1]
    with open("data/md/" + name + ".md", "w+", encoding='utf-8') as f:
        f.write(dictionary['markdown'])
```
